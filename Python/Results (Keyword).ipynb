# Variation of Execution time with input length: Serial

```python
import matplotlib.pyplot as plt

fig = plt.figure()
my_file = open("Serial_inputlen.txt", "r")
content = my_file.read()
content_list =content.split("\n")
y = list(map(float,content_list))
my_file.close()
plt.style.use('dark_background')
ax = fig.add_axes([0,0,1,1])

plt.ylabel('Time taken (in s)',fontsize=20)
plt.xlabel('Word length',fontsize=20)
plt.title('Variation in Execution time with Increasing Input Length - Serial\n Number of Words=100',fontsize=20)
x=[*range(100,1000000, 100000)]
plt.rcParams['figure.figsize'] = [8,8]
ax.plot(x,y,marker = 'o')
plt.show()
```

# Variation of Execution time with input length: Parallel

# Comparison of Serial vs Parallel with Increasing Word Length

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
my_file = open("Serial_inputlen.txt", "r")
content = my_file.read()
content_list =content.split("\n")
y = list(map(float,content_list))
my_file.close()
plt.style.use('dark_background')
ax = fig.add_axes([0,0,1,1])

plt.ylabel('Time taken (in s)',fontsize=20)
plt.xlabel('Number of words',fontsize=20)
x=[*range(100,1000000, 100000)]

ax.plot(x,y,marker = 'o',label='Serial')



my_file = open("parallel_input_len.txt", "r")
content = my_file.read()
content_list =content.split("\n")
y = list(map(float,content_list))
my_file.close()
plt.ylabel('Time taken (in s)',fontsize=18)
plt.xlabel('Word Length',fontsize=18)
x=[*range(100,1000000, 100000)]
plt.rcParams['figure.figsize'] = [8,8]
plt.title('Comparison Between Serial And Parallel \n Number of Words=100',fontsize=20)

ax.plot(x,y,marker = 'o',label="Parallel")
plt.legend(['Serial',"Parallel"], loc ="upper left")

plt.show()
```
